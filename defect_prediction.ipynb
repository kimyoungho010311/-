```python
import warnings
warnings.filterwarnings(action='ignore') # 경고를 인쇄하지 않음

import numpy as np
import scipy as sp
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('/Users/kim-youngho/Desktop/data_sience/uci-secome.csv')
```

```python
result = data['Pass/Fail']
```

```python
value_counts = result.value_counts()
# -1을 'pass'로, 다른 값을 'fail'로 매핑
value_counts.index = ['Pass' if x == -1 else 'Fail' for x in value_counts.index]
```

```python
result_pie_chart = value_counts.plot.pie(autopct='%.1f%%', startangle=90, figsize=(8, 6))
plt.title('Pass/Fail Distribution')
plt.ylabel('')  # y축 레이블 제거
```
